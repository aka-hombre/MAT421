##Cubic Spline Interpolation##

This method of interpolation uses a set of peicewise cubic functions. In other words, we asume $(x_i, y_i)$ and $(x_{i+1}, y_{i+1})$ are conected by a cubic:

$S_i(x)=a_ix^3+b_ix^2+c_ix+d$ which is valid for $x_i\le x \le x_{i+1}$ for $i=1,\ldots ,n-1$.


To solve this we need:


*   Find the coefficients values for each function *ie.* $a_i,b_i\ldots$ We need $1-n$ functions for $n$ points

*Below are examples of cubic interpolations, using the same data from the linear interpolations. Notice the how these functions are smoother and curved!*


```python
from scipy.interpolate import CubicSpline
import numpy as np
import matplotlib.pyplot as plt

x = [0, 1, 2, 3, 4]
y = [1, 3, 2, 4, 3]

f = CubicSpline(x, y, bc_type='natural')
x_new = np.linspace(0, 4, 100)
y_new = f(x_new)

plt.figure(figsize = (10,8))
plt.plot(x_new, y_new, 'b')
plt.plot(x, y, 'ro')
plt.title('Cubic Spline Interpolation')
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```

```python
from scipy.interpolate import CubicSpline
import numpy as np
import matplotlib.pyplot as plt

x=[]
y=[]

xvalue = 0.1

for i in range(0,10): #This generates some points to try our methods using a y=1/x
  x.append(xvalue)
  y.append((1/x[i]))
  xvalue += 0.1

f = CubicSpline(x, y, bc_type='natural')
x_new = np.linspace(0, 1, 100)
y_new = f(x_new)

plt.figure(figsize = (10,8))
plt.plot(x_new, y_new, 'b')
plt.plot(x, y, 'ro')
plt.title('Cubic Spline Interpolation of y=1/x')
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```
